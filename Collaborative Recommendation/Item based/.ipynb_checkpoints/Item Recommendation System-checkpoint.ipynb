```python
data=ratings.pivot_table(index='user_id',columns='movie_id',values='rating')
```

```python
data_matrix=data.replace(np.nan,0)
```

```python
data_matrix
```

```python
from sklearn.metrics.pairwise import pairwise_distances 
user_similarity = pairwise_distances(data_matrix, metric='cosine')
item_similarity = pairwise_distances(data_matrix.T, metric='cosine')
```

```python
item_similarity 
```

```python
def predict(ratings, similarity, type='user'):
    ratings = np.asarray(ratings)

    if type == 'user':
        mean_user_rating = ratings.mean(axis=1)

        ratings_diff = ratings - mean_user_rating[:, np.newaxis]

        pred = (
            mean_user_rating[:, np.newaxis]
            + similarity.dot(ratings_diff)
            / np.array([np.abs(similarity).sum(axis=1)]).T
        )

    elif type == 'item':
        pred = (
            ratings.dot(similarity)
            / np.array([np.abs(similarity).sum(axis=1)])
        )

    return pred
```

```python
user_prediction = predict(data_matrix, user_similarity, type='user')
item_prediction = predict(data_matrix, item_similarity, type='item')
```

```python
def itembased(input_item, item_similarity, item_predictions,
              similar_item_count, thres):

    input_index = input_item - 1

    item_sim_table = pd.DataFrame(item_similarity)

    similar_item_indexes = (
        item_sim_table[input_index]
        .sort_values(ascending=False)
        .iloc[1:similar_item_count + 1]
        .index
        .tolist()
    )

    similar_movie_ids = [
        index + 1
        for index in similar_item_indexes
    ]
    item_pred_df = pd.DataFrame(item_predictions)

    recommended_movie_ids = []

    for movie_id in similar_movie_ids:

        movie_index = movie_id - 1

        # Average predicted rating for this movie
        predicted_score = item_pred_df.iloc[:, movie_index].mean()

        if predicted_score >= thres:
            recommended_movie_ids.append(movie_id)
    i_cols = [
        'movie id',
        'movie title',
        'release date',
        'video release date',
        'IMDb URL',
        'unknown',
        'Action',
        'Adventure',
        'Animation',
        "Children's",
        'Comedy',
        'Crime',
        'Documentary',
        'Drama',
        'Fantasy',
        'Film-Noir',
        'Horror',
        'Musical',
        'Mystery',
        'Romance',
        'Sci-Fi',
        'Thriller',
        'War',
        'Western'
    ]

    final_recommend_movies = (
        items[
            items['movie id'].isin(recommended_movie_ids)
        ]['movie title']
        .tolist()
    )

    print("Input Movie ID:", input_item)

    print(
        "Similar Movie IDs:",
        similar_movie_ids
    )

    print(
        "Recommended Movie IDs:",
        recommended_movie_ids
    )

    print(
        "Recommended Movies:",
        final_recommend_movies
    )


    return final_recommend_movies
```
