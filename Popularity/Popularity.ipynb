```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating', 'timestamp']

ratings = pd.read_csv('ml-100k/u.data',sep='\t',names=r_cols,encoding='latin-1')

i_cols = [
    'movie_id', 'movie_title', 'release_date','video_release_date', 'IMDb_URL','unknown', 'Action', 'Adventure', 'Animation',"Children's", 'Comedy', 'Crime','Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical','Mystery', 'Romance', 'Sci-Fi','Thriller', 'War', 'Western']

movies = pd.read_csv('ml-100k/u.item',sep='|',names=i_cols, encoding='latin-1')

action_movies = movies[movies['Action'] == 1]

action_rating = pd.merge(ratings,action_movies,on='movie_id')

popular_action = action_rating.groupby( ['movie_id', 'movie_title'])['rating'].mean().reset_index()

popular_action.rename(columns={'rating': 'Average Rating'}, inplace=True)

top5 = popular_action.sort_values(by='Average Rating',ascending=False).head(5)

print("Top 5 Popular Action Movies")
print(top5)
```
